```python
class Store:
    def __init__(self):
        self.items = { }

    def stockup(self, item, count):
        self.items[item] = count if item not in self.items else self.items[item] + count

    def checkout(self, item, count):
        if item not in self.items:
            raise KeyError("Requested item not in the directory")
```

```python
import unittest

class TestStore(unittest.TestCase):
    def test_checkout(self):
        store = Store()
        store.stockup('Paracetamol', 25)
        with self.assertRaises(KeyError):
            store.checkout('Paracetamol', 30)

    def main(a):
        unittest.main()

TestStore.main(123)
```

```python
# Iterator allows you to traverse across a collection

# defined using __iter__ and __next__ methods implementation

class MyNumbers:
    def __iter__(self):
        self.num = 1
        return self

    def __next__(self):
        if self.num <= 5:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration

obj = MyNumbers()
myiter = iter(obj)

for i in myiter:
    print(i)
```

```python
class MyIterator:
    def __iter__(self):
        self.count = 0
        self.a = 0
        self.b = 1
        return self

    def __next__(self):
        self.count += 1
        temp = self.a
        self.a = self.b
        self.b += temp
        if self.count == 1:
            return temp
        if self.count == 2:
            return self.a
        if self.count > 10:
            raise StopIteration
        else:
            return temp

obj = MyIterator()
myiter = iter(obj)

for i in myiter:
    print(i)
```

```python
'''
Iterator requires class with iter and next
Generator uses function with yield keyword
Generator more memory efficient, Iterator more control
Iterator manual implementation, Generator consise implementation
'''

def my_gen():
    for i in range(5):
        yield i

gen = my_gen()
for val in gen:
    print(val)
```

```python
def fibo(limit):
    a, b = 1, 0
    for count in range(limit):
        c = a + b
        a = b
        b = c
        yield a

gen = fibo(10)
for val in gen:
    print(val)
```
