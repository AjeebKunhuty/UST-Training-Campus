```python
import logging

logging.basicConfig( level = logging.INFO)

while(True):
    try:    
        num1 = int(input("Enter number 1"))
        num2 = int(input("Enter number 2"))

        print("""
            + for addition
            - for subtraction
            * for multiplication
            / for division
            X for exit
        """)
        choice = input("Enter choice")
        if choice == '+':
            result = num1 + num2
        elif choice == '-':
            result = num1 - num2
        elif choice == '*':
            result = num1 * num2
        elif choice == '/':
            result = num1 / num2
        elif choice == 'X':
            break

        print(result)
        
    except ValueError:
        logging.error(" Enter integer values")
    except ZeroDivisionError:
        logging.error(" Number 2 cannot be zero for division")

print("Exited calculator")
```
