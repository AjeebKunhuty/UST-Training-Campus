```python
# Assignment 2

# Where to use map, reduce and filter

# Map can be used when list of names need to be titled
def caps(name):
    return name.title()

names = list(map(caps, ["tony stark", "steve jobs","krishna sudheendra"]))
# print(names)

# Filters can be used to filter out names that do not have the letter 'A' in it

filteredNames = list(filter(lambda name : not ('a' in name or 'A' in name), names))
# print(filteredNames)

# Reduce can be used for getting the total amount from payrolls of all employees
from functools import reduce
payroll = [1000,5000,20000]
total = reduce(lambda x, y: x + y, payroll)
# print(total)
```
