```python
# Assignment 1 : matrix transpose without using functions
"""
    Find the transpose of a matrix given as list of lists
"""
matrix = [
    [1,2,3,4],
    [6,8,3,4],
    [7,8,9,3]
]

transposed = [ [] for i in range(len(matrix[0]))]
for i in range(len(matrix)):
    for j in range(len(transposed)):
        transposed[j].append(matrix[i][j])
print(transposed)
```

```python
# using List comprehension
transposed = [ [ matrix[j][i] for j in range(len(matrix)) ] for i in range(len(matrix[0])) ]
print(transposed)
```
