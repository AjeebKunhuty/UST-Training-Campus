```python
# function for hcf

def hcf(num1, num2):
    """ HCF of two numbers """
    start = min(num1, num2)
    for factor in range(start, 0, -1):
        if num1 % factor == 0 and num2 % factor == 0:
            return factor
    return 1

print(hcf(20, 21))
```

```python
def hcf(num1, num2):
    factors = [factor for factor in range( min(num1, num2), 0, -1) if num1 % factor == 0 and num2 % factor == 0 ]
    return max(factors)

print(hcf(30, 75))
```

```python
class Sports:
    def __init__(self, model, color, cost):
        self.__model = model
        self.__color = color
        self.__cost = cost

    def getModel(self):
        return self.__model

    def setModel(self, model):
        self.__model = model

    def getColor(self):
        return self.__color
        
    def setColor(self, color):
        self.__color = color

    def getCost(self):
        return self.__cost
        
    def setCost(self, cost):
        self.__cost = cost

class Shoes(Sports):
    def __init__(self, model, color, cost, brand):
        super().__init__(model, color, cost)
        self.__brand = brand

    def getInfo(self):
        print(f"{self.__brand}, {self.getModel()}, {self.getColor()}, {self.getCost()}")

shoe = Shoes("AC300", "Black", 5000, "Nike")
shoe.getInfo()
```
